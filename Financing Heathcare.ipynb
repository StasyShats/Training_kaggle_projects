```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('annual-healthcare-expenditure-per-capita.csv')
other = pd.read_csv('life-expectancy-vs-healthcare-expenditure.csv')
```

```python
other.head()
```

```python
other = other.drop(columns= ['Entity', 'Current health expenditure per capita, PPP (current international $)', 'Continent'])
other.head()
```

```python
other.shape[0]
```

```python
other= other.dropna(axis='rows')
other.shape[0]
```

```python
df.shape
```

```python
df = df.dropna(axis='rows')
df.shape[0]
```

```python
df = df.merge(other, how='inner', on=['Code', 'Year'])
```

```python
df
```

```python
USA_Yearly_Healthcare_Spending = df[df['Code']=='USA']
USA_Yearly_Healthcare_Spending
```

```python
plt.scatter(df['Indicator:Current health expenditure (CHE) per capita in PPP int$'], df['Life expectancy at birth, total (years)'], color='b')
plt.scatter(USA_Yearly_Healthcare_Spending['Indicator:Current health expenditure (CHE) per capita in PPP int$'], USA_Yearly_Healthcare_Spending['Life expectancy at birth, total (years)'], color='r')
plt.xlabel('Health expendeture per capita')
plt.ylabel('Life expectancy')
plt.title('Health expendeture vs life expentancy')
```
